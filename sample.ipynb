# IDA CFG Analysis with NetworkX

This notebook demonstrates how to extract, load, and visualize the Control Flow Graph (CFG) from an IDA database.


### 1. Extract CFG from IDA Database
This step requires `idapro` 9.0+ and a valid license.


```python
import sys
import os
from src.extract_cfg_standalone import extract_cfg_from_db
import os

# Replace with the path to your IDA database (.idb or .i64)
db_path = "/Users/mark/windows_share/test/reorder_and_pad.exe.i64"
json_path = "reorder_and_pad2.json"

if os.path.exists(db_path):
    json_path = extract_cfg_from_db(db_path, output_path=json_path)
else:
    print(f"Database {db_path} not found. Skipping extraction.")
```

### 2. Load and Visualize the Graph


```python
import networkx as nx
import matplotlib.pyplot as plt
from src.visualize_cfg import load_cfg
import os

# Load the graph
G = load_cfg(json_path)

if G:
    print(f"Graph loaded: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

    # Basic info
    print("\nFunctions in the graph:")
    funcs = set(nx.get_node_attributes(G, 'func').values())
    for f in sorted(list(funcs)):
        print(f"- {f}")

    # Visualize (only if small enough)
    if G.number_of_nodes() < 50:
        plt.figure(figsize=(12, 8))
        pos = nx.spring_layout(G)
        labels = nx.get_node_attributes(G, 'label')
        nx.draw(G, pos, labels=labels, with_labels=True, node_size=2000, node_color="skyblue", font_size=8, arrowsize=20)
        plt.title("IDA Control Flow Graph")
        plt.show()
    else:
        print("\nGraph is too large for simple visualization. Use NetworkX analysis functions.")

    # Example Analysis: Find nodes with high degree
    degrees = dict(G.degree())
    top_nodes = sorted(degrees.items(), key=lambda x: x[1], reverse=True)[:5]
    print("\nTop 5 nodes by degree:")
    for node, degree in top_nodes:
        label = G.nodes[node].get('label', hex(node))
        print(f"Node {label}: {degree} edges")
else:
    print(f"Could not load graph from {json_path}. Please make sure the file exists and it is a valid CFG JSON.")
```
